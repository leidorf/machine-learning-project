```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# İlk CSV dosyasının yolu
csv_file_path1 = '../data/cleanedearthquake.csv'

# İkinci CSV dosyasının yolu
csv_file_path2 = '../data/duzeltilmis_veriler.csv'

# İlk CSV dosyasını oku
df1 = pd.read_csv(csv_file_path1)

# İkinci CSV dosyasını oku
df2 = pd.read_csv(csv_file_path2)

# İki DataFrame'i birleştir
merged_df = pd.merge(df1, df2, how='outer')

# Tekrar eden satırları sil
merged_df = merged_df.drop_duplicates()

# Yeni bir CSV dosyasına kaydet
merged_csv_file_path = '../data/birlestirilmis_veriler.csv'
merged_df.to_csv(merged_csv_file_path, index=False)
```

```python
df = pd.read_csv(merged_csv_file_path)
```

```python
plt.figure(figsize=(16,16))
plt.subplot(2,1,1)
sns.countplot(x='Yer',data = df,order = df['Yer'].value_counts().index)
plt.xticks(rotation = 90)
plt.show()
```
